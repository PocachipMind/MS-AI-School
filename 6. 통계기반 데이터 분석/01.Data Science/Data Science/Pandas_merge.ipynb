```python
import pandas as pd
```

```python
# df1 df2 합치는 기능이있습니다.
df1 = pd.DataFrame({
    '날짜' : [1101,1102,1103,1104,1105],
    '온도' : [20, 18, 16, 14, 5]
}, columns=['날짜','온도'])
```

```python
df1
```

```python
df2 = pd.DataFrame(data = {
    '날짜' : [1101,1102,1103,1104,1109,1120],
    '습도' : [10, 20, 40, 100, 0, 50]
}, columns=['날짜','습도'])
# data라고 명시 안해도 첫 인자는 data임
```

```python
df2
```

```python
pd.merge(df1,df2) # 중복되는 부분만 합체 ( how = 'inner' 가 디폴트값임)
```

```python
pd.merge(df1, df2, how='outer') # 다 합쳐버리기
```

```python
pd.merge(df1, df2, how='left') # 온도 기준(왼쪽 데이터)으로 붙여버리기
```

```python
pd.merge(df1, df2, how='right')
```

```python
df3 = pd.DataFrame({
    'date' : [1101,1102,1103,1104,1105],
    '온도' : [20, 18, 16, 14, 5]
}, columns=['date','온도'])
df4 = pd.DataFrame(data = {
    '날짜' : [1101,1102,1103,1104,1109,1120],
    '습도' : [10, 20, 40, 100, 0, 50]
}, columns=['날짜','습도'])
```

```python
pd.merge(df3,df4,left_on='date',right_on='날짜')
```
