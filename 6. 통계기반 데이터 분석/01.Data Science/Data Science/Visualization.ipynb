# 바 그래프

```python
data = {
    '바나나' : 3, '사과' : 2, '오랜지' :10
}
# 선 : plot , 바 : bar
# print(data.keys(), data.values())
plt.bar([0,1,2], data.values())
plt.title('과일 개수')
plt.xticks([0, 1, 2], data.keys())
plt.yticks(sorted(data.values()))
plt.grid(True)
plt.xlabel('과일')
plt.ylabel('개수')
plt.show()
```

```python
data = {
    '바나나' : 3, '사과' : 2, '오랜지' :10
}
# 선 : plot , 세로바 bar, 가로바 barh
# print(data.keys(), data.values())
# x 값 y값 변경
plt.barh([0,1,2], data.values())
plt.title('과일 개수')
plt.yticks([0, 1, 2], data.keys())
plt.xticks(sorted(data.values()))
plt.grid(True)
plt.ylabel('과일')
plt.xlabel('개수')
plt.show()
```

# 파이차트
순서형 데이터가 아니라 카테고리에 많이 적용되는 차트

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Graphs']
explode = [0.05, 0.1, 0.05, 0.05]
colors = ['silver', 'gold','whitesmoke','lightgray']
# 파이차트 메서드 plt.pie
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, shadow=True, explode = explode, colors=colors)
plt.show()
```

```python
wedgeprops = {'width':0.7, 'linewidth':5}
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=0,\
        counterclock=True, colors=colors, wedgeprops=wedgeprops)
```

# 히스토그램

```python
heights = [168, 181, 164, 156, 178, 174, 161, 177, 166, 168, 159, 171, 180, 159, 167, 181, 169, 173, 169, 174, 170, 165]

# bins 는 계층의 수
plt.hist(heights, bins=10, label='heights')
plt.show()
```

```python
# 누적 히스토그램
plt.hist(heights, bins=10, label='heights', cumulative=True)
plt.show()
```

```python
plt.hist(heights, bins=10, label='heights', histtype='step')
plt.show()
```

```python
# density = 밀도로 나타냄 - 넓이 다 더하면 1임
plt.hist(heights, bins=10, label='heights', density=True)
plt.show()
```

```python
# alpha = 투명도
plt.hist(heights, bins=10, label='heights', alpha = 0.5, color='red')
plt.show()
```

# Scatter plot

```python
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x, y, c = colors, cmap='Spectral', s = area)
plt.colorbar()
plt.show()
```

# 박스 그래프

```python
np.random.seed(0)

plt.style.use('default')

# 정규 분포 하나 만들기
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

fig, ax = plt.subplots()

# 박스 그래프 : boxplot
ax.boxplot([data_a, data_b, data_c], notch= True, whis=1.5, vert=True)
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

```python
# 가로 박스 그래프
fig, ax = plt.subplots()
ax.boxplot([data_a, data_b, data_c], notch=True, whis=1.5, vert=False)
ax.set_xlim(-10.0, 10.0)
# ax.set_xlabel('Data Type')
# ax.set_ylabel('Value')
plt.show()
```

# 바이올린 플롯(Violin Plot)

```python
plt.style.use('default')

# 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 바이올린 : plt.violinplot
# showmeans - 평균값 표시, showextrema - 최소 최대 직선으로 표시, showmedians- 데이터 중앙값 표시
violin = plt.violinplot([data_a, data_b, data_c], showmeans=True, showextrema=True, showmedians= True)

plt.xticks(np.arange(1,4),labels=['A','B','C'])
plt.show()
```

```python
plt.style.use('default')

# 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 바이올린 : plt.violinplot
# showmeans - 평균값 표시, showextrema - 최소 최대 직선으로 표시, showmedians- 데이터 중앙값 표시
violin = plt.violinplot([data_a, data_b, data_c], showmeans=True, showextrema=True, showmedians= True)

plt.xticks(np.arange(1,4),labels=['A','B','C'])
violin['bodies'][0].set_facecolor('red')
violin['cbars'].set_edgecolor('red')
violin['cmaxes'].set_edgecolor('orange')
violin['cmins'].set_edgecolor('black')
violin['cmeans'].set_edgecolor('red')
violin['cmedians'].set_edgecolor('black')
plt.show()
```
