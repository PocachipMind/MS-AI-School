```python
import requests # 통신
import bs4 # 통신된 데이터 파악
```

```python
url = 'https://search.shopping.naver.com/search/category/100005307'
result = requests.get(url).text
result
```

```python
bsObj = bs4.BeautifulSoup(result, 'html.parser')
```

```python
items = bsObj.find_all('a',{'class':'basicList_link__JLQJf'})
```

```python
print(items[0].text)
print(items[1].text)
print(items[2].text)
print(items[3].text)
```

```python
lists = items[:3]
for l in lists:
  print(l.text)
```
